```python
# The first albumm run
import re

#Cleaning the lyric words from wrong format. In this exact order to of loops to make sure that there are no accidental joinings of words or skips because of capital letters.
def normalize_text(d):
    for key in d:
        if isinstance(d[key], str):
            d[key] = d[key].lower().replace('\n', ' ').replace('[chorus]', '').replace('[verse]', '').replace('[bridge]', '').replace('?','').replace(',','').replace('.','').replace('"','')

#This is the function for splitting all values inside the dictionary for each key
def split_values(e):
    for key in (e):
            e[key] = e[key].split()

#keeping these variable changes in new dictionaries to have data in the future
normalized_1979 = dict(numan_lyrics_1979)
normalize_text(normalized_1979)
split_values(normalized_1979)

split_lyrics_1979 = []
for key in normalized_1979:
    split_lyrics_1979.extend(normalized_1979[key])

#for 2017
normalized_2017 = dict(numan_lyrics_2017)
normalize_text(normalized_2017)
split_values(normalized_2017)

split_lyrics_2017 = []
for key in normalized_2017:
    split_lyrics_2017.extend(normalized_2017[key])

#print(split_lyrics_1979)
#print(split_lyrics_2017)
```

```python
# to find the most frequent words I found on google to use collections
import collections

#calculating the most frequent words. also made function here with 2nd variable where you enter the year for prints. Also saving the dictionaries of counter for future use in the wordcloud.
def counting_words_list(list, print_year):
    word_count = collections.Counter(list)
    top_10_words = word_count.most_common(10)
    print('TOP 10 MOST FREQUENT WORDS')
    print(f"Top 10 most frequently used words in {print_year}:")
    print('Occurrences: word')
    for word, count in top_10_words:
        print(f"{count}: {word}")
    print()
    return word_count

word_count = counting_words_list(split_lyrics_1979, 1979)
word_count_1979 = dict(word_count)

word_count = counting_words_list(split_lyrics_2017, 2017)
word_count_2017 = dict(word_count)
```

```python
#calculating the token ratio
def token_ratio(list, print_year):
    unique_words = set(list)
    total_words = len(list)
    token_ratio = len(unique_words) / total_words
    print(f"Ratio of unique words to total words in {print_year}: {token_ratio:.2f}")
    
print('UNIQUE WORD TO TOTAL WORDS (Higher meaning more unique words)')
token_ratio(split_lyrics_1979, 1979)
token_ratio(split_lyrics_2017, 2017)
```

```python
#calculating the average word lenght
def average_word_lenght(list, print_year):
    total_characters = sum(len(word1) for word1 in list)
    total_words = len(list)
    avg_word_length = total_characters / total_words
    
    print(f"Average word length in {print_year}: {avg_word_length:.2f} characers")

print('AVERAGE WORD LENGHT')
average_word_lenght(split_lyrics_1979, 1979)
average_word_lenght(split_lyrics_2017, 2017)
```

```python
#calculating the longest and shortest songs. Here I also developed a code that joins string to be in ""
def longest_shortest_songs(normalized, print_year):
    lenghts = {}
    for key in normalized.keys():
        lenghts[key] = len(normalized[key])
    max_key = []
    min_key = []
    #print(lenghts)
    for key, value in lenghts.items():
        if value == max(lenghts.values()):
            max_key.append(key)
    for key, value in lenghts.items():
        if value == min(lenghts.values()):
            min_key.append(key)
    min_key_str = '", and "'.join(min_key)
    max_key_str = '", and "'.join(max_key)
    print(f'The highest amount of words in the {print_year} album has the song "{max_key_str}" which is equal to {max(lenghts.values())} words')
    print(f'The lowest amount of words in {print_year} is in "{min_key_str}" which is equal to {min(lenghts.values())} words')
    print()
    
longest_shortest_songs(normalized_1979, 1979)
longest_shortest_songs(normalized_2017, 2017)
```
